```python
s = """			<!-- XX -->
			<section id="aXXX" class="spotlight styleXXXX bottom">
				<span class="image fit main bottom"><img src="images/picXXXXX.png" alt="" /></span>
				<div class="content">
					<header>
						<h2>Title</h2>
						<p>Subtitle</p>
					</header>
					<p>Paragraph.</p>
				</div>
				<a href="#aXXXXXX" class="goto-next scrolly">Next</a>
			</section>"""

from num2words import num2words as n2w

def Format(x):
    x = str(x)
    return "0"*(2-len(x)) + x

for i in range(5, 15):
    t = s.replace("XXXXXX", n2w(i+1))
    t = t.replace("XXXXX", Format(i+1))
    t = t.replace("XXXX", str(((i-1)%3)+1))
    t = t.replace("XXX", n2w(i))
    t = t.replace("XX", n2w(i).capitalize())

    print(t)
    print()
```
